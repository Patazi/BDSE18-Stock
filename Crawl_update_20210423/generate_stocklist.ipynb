# 台股號碼一覽表-上市

```python
import pandas as pd
df = pd.read_csv('2.csv', names=['首列','有價證券代號及名稱','國際證券辨識號碼(ISIN Code)','上市日','市場別','產業別','CFICode','備註'])
```

```python
df.head(3) #看看編碼有沒有錯誤
```

```python
df.info() #看看資料型態
```

```python
mask = df["CFICode"] == "ESVUFR" #篩選出台股
df[mask]
```

```python
df2 = df[mask].reset_index(drop=True)
df2[['有價證券代號及名稱','國際證券辨識號碼(ISIN Code)','上市日','市場別','產業別','CFICode']]
```

```python
df2['有價證券代號'] = df2['有價證券代號及名稱'].str.split('\s+').str.get(0)
df2['名稱'] = df2['有價證券代號及名稱'].str.split('\s+').str.get(1)
df2["上市日"] = pd.to_datetime(df2["上市日"], format='%Y/%m/%d')
df3 = df2[['有價證券代號','名稱','國際證券辨識號碼(ISIN Code)','上市日','市場別','產業別','CFICode']]
df3
#這段執行下來有 warning 出現，<ipython-input-212-cf988f8a1199>:2: SettingWithCopyWarning: 
# A value is trying to be set on a copy of a slice from a DataFrame.
# Try using .loc[row_indexer,col_indexer] = value instead
#應該可以再把 code 寫的更好
```

# 台股號碼一覽表-上櫃

```python
df5 = df4[mask].reset_index(drop=True)
df5[['有價證券代號及名稱','國際證券辨識號碼(ISIN Code)','上市日','市場別','產業別','CFICode']]
```
