# 除權息日程表

```python
def DividendResult(stockNo,Name):
#     try:
    empty_list=[]
    import time
    from datetime import datetime
    import requests
    import pandas as pd
    starttime=datetime.now()#紀錄爬取該支股票要多久
    url = "https://api.finmindtrade.com/api/v4/data"
    parameter = {
        "dataset": "TaiwanStockDividendResult",
        "data_id": stockNo,
        "start_date": "2008-01-01",
        "token": "", # 參考登入，獲取金鑰
    }
    data = requests.get(url, params=parameter)
    data = data.json()
    data = pd.DataFrame(data['data'])
    if data.empty != True:
        # print(data)
        # print(data.info())
        data.insert(0,column='stock_name',value=Name)
        data = data[['stock_id','stock_name','date','before_price','after_price','stock_and_cache_dividend','stock_or_cache_dividend','max_price','min_price','open_price','reference_price']]

        file_name = "DividendResult/{}_DividendResult.csv".format(stockNo)
        data.to_csv(file_name, index=False)
        print(file_name+"下載完成...", end="")
        endtime= datetime.now()#紀錄爬取該支股票要多久
        print("執行時間:",endtime-starttime,"秒")#紀錄爬取該支股票要多久
    #     except:
    #         print(stockNo+" "+Name+"有問題")
    else:
        print(stockNo+"為空")
```

```python
import csv
# with open('stock_list_for_test.csv') as f:#供小型測試用
with open('stock_list.csv') as f:
    list_of_stock = csv.DictReader(f)
    for obj in list_of_stock:
        DividendResult(obj['stock_id'],obj['stock_name'])
```
